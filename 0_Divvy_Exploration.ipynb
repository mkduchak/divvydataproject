## 1.2 Weather Data

```python
weather = pd.read_csv('/home/michael/Documents/Projects/divvydataproject/data/weatherhistory.csv')
weather.head()
```

```python
weather.dtypes
```

```python
weatherdata = weather[['STATION_NAME','DATE','HOURLYDRYBULBTEMPF','HOURLYPrecip']].copy(deep = True)
weatherdata['DATE'] = pd.to_datetime(weatherdata['DATE'])
weatherdata = weatherdata[pd.notnull(weatherdata['HOURLYPrecip'])]
weatherdata['datetime'] = weatherdata['DATE'].dt.round('H')


weatherdata['HOURLYDRYBULBTEMPF'] = pd.to_numeric(weatherdata['HOURLYDRYBULBTEMPF'], errors = 'coerce')

weatherdata['HOURLYPrecip'] = [x.replace('T', '').replace('s','') for x in weatherdata['HOURLYPrecip']]
wdcleanup = []
for x in weatherdata['HOURLYPrecip']:
    try:
        new = float(x)
        wdcleanup.append(new)
    except:
        wdcleanup.append(0)
        

weatherdata['HOURLYPrecip'] = wdcleanup
weatherdata.tail(10)
```

```python
weatherdata.groupby('HOURLYDRYBULBTEMPF', as_index = False)[['DATE']].count()
```

```python
weatherdata.to_pickle('/home/michael/Documents/Projects/divvydataproject/data/weatherdataclean.pkl')
```

## 1.4 Park Location Data

```python
parks = pd.read_csv('/home/michael/Documents/Projects/divvydataproject/data/Parks_-_Locations__deprecated_November_2016_.csv')
```

```python
parks.head()
```

```python
parks.columns
```

```python
parks.iloc[5]['LOCATION']
```

```python
lookup = re.compile('\(([^)]+)\)')
location = lookup.findall(parks.iloc[8]['LOCATION'])[0]
location
```

```python
parks['location'] = parks['LOCATION'].str.extract(lookup, expand = True)
```

```python
parks.head()
```

## 1.5 Assemble lookup table

```python
parkslookup = parks[['PARK NUMBER','location']].copy(deep=True)
parkslookup.columns = ['id','location']
parkslookup['id'] = 'p' + parkslookup['id'].astype(str)
parkslookup = parkslookup[pd.notnull(parkslookup['location'])]
```

```python
parkslookup.dtypes
```

```python
parkslookup['latitude'] = [float(str(x).split(',')[0]) for x in parkslookup['location']]
parkslookup['longitude'] = [float(str(x).split(',')[1]) for x in parkslookup['location']]
```

```python
parkslookup = parkslookup[['id','latitude','longitude']]
```

```python
parkslookup.head()
```

```python
stationlookup = stations[['id','latitude','longitude']]
```

```python
stationlookup.tail()
```

```python
distlookup = pd.concat([stationlookup, parkslookup])
```
